```python
import pywindow as pw
```

```python
# Save list of xyz file names 
xyzName = os.listdir('./Structures/')
```

```python
def list_append(id, xyzName, totalproList):
    for molec in tqdm(xyzName):
        molsys = pw.MolecularSystem.load_file(f'./Structures/{molec}')
        molec = molec.strip('.xyz')                      
        mol = molsys.system_to_molecule()
        centrMass = mol.calculate_centre_of_mass()
        maxDiam = mol.calculate_maximum_diameter()
        aveDiam = mol.calculate_average_diameter()
        poreDiam = mol.calculate_pore_diameter()
        poreVol = mol.calculate_pore_volume()
        poreDiamOpt = mol.calculate_pore_diameter_opt()
        poreVolOpt = mol.calculate_pore_volume_opt()

        molPropList = [molec, centrMass[0], centrMass[1], centrMass[2], maxDiam, aveDiam, poreDiam, poreVol, poreDiamOpt, poreVolOpt]

        totalproList.append(molPropList)
    
    #return totalproList
```

```python
totalproList = list()
```

```python
if __name__ == "__main__":
    proc = 25
    jobs = []
    for i in range(0, proc):
        totalproList = list()
        #list_append(molec, totalproList)
        process = multiprocessing.Process(target=list_append,
                                          args=(i, xyzName, totalproList))
        jobs.append(process)
        
    # Start the processes (i.e. calculate the random number lists)
    for j in jobs:
        print(f'start {j} job')
        
        j.start()
    
    # Ensure all of the processes have finished
    for j in jobs:
        print(f'start compiling {j} results')
        j.join()

    print("List processing complete.")
```
