```python
#!conda install -y -c openbabel openbabel 
import openbabel
from tqdm import tqdm_notebook as tqdm
import math
```

```python
# Read Files 
train = pd.read_csv('train.csv')
test = pd.read_csv(f'test.csv')
structures = pd.read_csv('structures.csv')
x = structures.groupby('molecule_name').atom_index.max().reset_index(drop=False)
x.columns = ['molecule_name','totalatoms']
x.totalatoms+=1
train = train.merge(x,on='molecule_name')
test = test.merge(x,on='molecule_name')
#train = train[train.molecule_name=='dsgdb9nsd_000001']
```

```python
obConversion = openbabel.OBConversion()
obConversion.SetInFormat("xyz")
structdir='./Structures/'
mols=[]
mols_files=os.listdir(structdir)
mols_index=dict(map(reversed,enumerate(mols_files)))
for f in mols_index.keys():
    try:
        mol = openbabel.OBMol()
        obConversion.ReadFile(mol, structdir+f) 
        mols.append(mol)
    except:
        print("Wrong format")
        pass
```

```python
# Get atom information for train
def create_atom_features_ob(df, mols):
    stats = []
    for m, groupdf in tqdm(df.groupby('molecule_name')):
        mol=mols[mols_index[m+'.xyz']]
        for i in groupdf.index.values:
            totalatoms = groupdf.loc[i].totalatoms
            firstatomid = int(groupdf.loc[i].atom_index_0)
            secondatomid = int(groupdf.loc[i].atom_index_1)
            entrystats = {}
            entrystats['totalatoms'] = totalatoms
            #entrystats['scalar_coupling_constant'] = float(groupdf.loc[i].scalar_coupling_constant)
            entrystats['type'] = groupdf.loc[i]['type']
            a = mol.GetAtomById(firstatomid)
            b = mol.GetAtomById(secondatomid)
            entrystats['molecule_name'] = m
            entrystats['atom_index_0'] = firstatomid
            entrystats['atom_index_1'] = secondatomid
            entrystats['bond_distance'] = a.GetDistance(b)
            entrystats['bond_atom'] = b.GetType()
            entrystats['atom_0_mass'] = a.GetAtomicMass()
            entrystats['atom_1_mass'] = b.GetAtomicMass()
            entrystats['atom_0_valence'] = a.GetValence()
            entrystats['atom_1_valence'] = b.GetValence()
            entrystats['atom_0_hybrid'] = a.GetHyb()
            entrystats['atom_1_hybrid'] = b.GetHyb()
            entrystats['atom_0_partcharge'] = a.GetPartialCharge()
            entrystats['atom_1_partcharge'] = b.GetPartialCharge()
            #entrystats['atom_0_residue'] = a.GetResidue()
            #entrystats['atom_1_residue'] = b.GetResidue()

            #Put the tertiary data in order of distance from first hydrogen
            tertiarystats = {}
            for j,c in enumerate(list(set(range(totalatoms)).difference(set([firstatomid,secondatomid])))):
                tertiaryatom = mol.GetAtomById(c)
                tp = tertiaryatom.GetType()
                dist = a.GetDistance(tertiaryatom)
                ang = a.GetAngle(b,tertiaryatom)*math.pi/180
                partcharge = tertiaryatom.GetPartialCharge()
                while(dist in tertiarystats):
                    dist += 1e-15
                       # print('Duplicates!',m,j,dist)
                tertiarystats[dist] = [tp,dist,ang, partcharge]#

            for k, c in enumerate(sorted(tertiarystats.keys())):
                entrystats['tertiary_atom_'+str(k)] = tertiarystats[c][0]
                entrystats['tertiary_distance_'+str(k)] = tertiarystats[c][1]
                entrystats['tertiary_angle_'+str(k)] = tertiarystats[c][2]
                entrystats['tertiary_partcharge_'+str(k)] = tertiarystats[c][3]
                
            stats.append(entrystats)
    gc.collect()
    
    outfile = pd.DataFrame(stats)
    #if 'scalar_coupling_constant' in outfile.columns:
    #    outfile = outfile.drop(['scalar_coupling_constant'],  axis=1)
    
    return outfile
```

```python
# Get Molecule information train
def create_molec_features_ob(df, mols):
    stats = []
    for m, groupdf in tqdm(df.groupby('molecule_name')):
        mol=mols[mols_index[m+'.xyz']]
        entrystats = {}
        entrystats['molecule_name'] = m
        entrystats['molecule_energy'] = mol.GetEnergy()
        entrystats['molecule_wt'] = mol.GetMolWt()
        entrystats['molecule_exact_mass'] = mol.GetExactMass()
        entrystats['molecule_total_charge'] = mol.GetTotalCharge()
        entrystats['molecule_total_spin_mult'] = mol.GetTotalSpinMultiplicity()
        entrystats['molecule_dimension'] = mol.GetDimension()
        entrystats['molecule_compute_charge'] = openbabel.OBChargeModel.ComputeCharges(mol)
        entrystats['molecule_dipole_moment_x'] = openbabel.OBChargeModel.GetDipoleMoment(mol)[0]
        entrystats['molecule_dipole_moment_y'] = openbabel.OBChargeModel.GetDipoleMoment(mol)[1]
        entrystats['molecule_dipole_moment_z'] = openbabel.OBChargeModel.GetDipoleMoment(mol)[2]
        
        stats.append(entrystats)
        
    outfile = pd.DataFrame(stats)
    return outfile
    
```

```python
df_mol_train = create_molec_features_ob(df = train, mols = mols)
```

```python
df_mol_test = create_molec_features_ob(df = test, mols = mols)
```
